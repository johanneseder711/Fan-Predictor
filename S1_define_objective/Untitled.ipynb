# Exploring Regression Metrics

```python
import numpy as np
from random import random
import matplotlib.pyplot as plt
```

```python
# create sample arrays
y_true = np.array([123, 264, 77, 189])
y_pred = np.array([120, 250, 88, 195])
```

```python
def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))
```

```python
mae(y_true, y_pred)
```

```python
def mse(y_true, y_pred):
    return np.mean(np.square(y_true - y_pred))
```

```python
mse(y_true, y_pred)
```

```python
def huber(y_true, y_pred, delta = 1):
    return np.mean([0.5*(y_true[i] - y_pred[i])**2 if (np.abs(y_true[i] - y_pred[i]) <= delta) else delta*(np.abs(y_true[i] - y_pred[i]) - 0.5*delta**2) for i in range(len(y_true)) ])
```

```python
huber(y_true, y_pred)
```

```python
delta = 5
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,7))
ax_1 = ax[0]
x_vals = np.arange(-100,100,0.01)
ax_1.plot(x_vals, np.abs(x_vals), label='mae')
ax_1.plot(x_vals, np.square(x_vals), label='mse')
ax_1.plot(x_vals, np.where(np.abs(x_vals)<delta, 0.5*np.square(x_vals), delta*(np.abs(x_vals) - 0.5*delta)), label='huber')
ax_1.legend(prop={'size': 15})
ax_1.set_title('comparison of losses', fontsize=15)
ax_1.set_ylim(0,100)
ax_1.grid(True)

ax_2 = ax[1]
x_vals = np.arange(-10,10,0.01)
ax_2.plot(x_vals, np.abs(x_vals), label='mae')
ax_2.plot(x_vals, np.square(x_vals), label='mse')
ax_2.plot(x_vals, np.where(np.abs(x_vals)<=delta, 0.5*np.square(x_vals), delta*(np.abs(x_vals) - 0.5*delta)), label='huber')
ax_2.legend(prop={'size': 15})
ax_2.vlines(-delta, 0,100,color='red',linestyles='--')
ax_2.vlines(delta, 0,100,color='red',linestyles='--')
ax_2.set_title('comparison of zoomed losses', fontsize=15)
ax_2.grid(True)


plt.show()
```

```python
mse(y_true, y_pred)
```

```python
huber(y_true, y_pred, delta=1.5)
```
